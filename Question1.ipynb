Part 1:

The chickwts.txt file contains two columns "weight" and "feed." These columns have 72 rows with the weight of the feed component. The feed components include horsebean, linseed, soybean, sunflower, meatmeal, and casein. 

Part 2: 

Plot of data from the chickwts.txt file:

```python
import pandas
import numpy
from plotnine import *

data=pandas.read_csv("chickwts.txt",sep=',')
plot=ggplot(data)+theme_classic()+xlab("feed")+ylab("weight")
x_ax="factor(feed)"
y_ax=data['weight']

print plot + geom_bar(aes(x=x_ax),y=y_ax)
```
